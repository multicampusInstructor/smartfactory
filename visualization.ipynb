```python
with open('./data/Abc1115.csv','r') as file:
    lines = file.readlines()
student = []    
for line in lines:
    temp = line[:-1].split(',')
    for idx,i in enumerate(temp[:8]):
        i = i.strip()
        if i.isdigit():
            temp[idx] = int(i)
    student.append(temp)
```

```python
import matplotlib.pyplot as plt
```

```python
year = ['2018','2019','2020','2021']
sale = [342,543,256,457]

plt.figure(figsize=(4,5))
plt.bar(year,sale)
plt.xlabel('year')
plt.ylabel('sale')
# plt.title('test')
for i, v in enumerate(year):
    plt.text(v, sale[i], sale[i],                
             fontsize = 9, 
             color='blue',
             horizontalalignment='center',  
             verticalalignment='bottom')
```

```python
plt.plot(['2016','2017','2018','2019'],[140,69,105,220])
plt.show()
```

```python
import time
for sub in range(3,8):
    data = []
    for temp in student:
        data.append(temp[sub])
    plt.hist(data)
    plt.show()
    time.sleep(10)
```

```python
abc = {'width':0.7,'edgecolor':'w', 'linewidth':0}
plt.pie(sale,labels=year,autopct='%0.1f%%',startangle=0,
        counterclock=False,explode=[0,0.1,0,0],shadow=False,wedgeprops=abc)
plt.show()
```

```python
import numpy as np
a = [23,45,43]
b = [13,35,83]
c = [38,15,83]
year1 = year = ['2019','2020','2021']
index = np.arange(3)
bar_width = 0.25
plt.figure(figsize=(4,3))

b1 = plt.bar(index, a, bar_width, alpha=0.4, color='red', label='a')

b2 = plt.bar(index + bar_width, b, bar_width, alpha=0.4, color='blue', label='b')

b3 = plt.bar(index + 2 * bar_width, c, bar_width, alpha=0.4, color='green', label='c')

# x축 위치를 정 가운데로 조정하고 x축의 텍스트를 year 정보와 매칭
plt.xticks(np.arange(bar_width, 3 + bar_width, 1), year)

# x축, y축 이름 및 범례 설정
plt.xlabel('년도', size = 13)
plt.ylabel('revenue', size = 13)
plt.legend()
plt.show()
```
